```python
ratings_games = ratings.merge(games, on = "Game ID")
```

```python
number_of_ratings = ratings_games.groupby("name")["Review Score"].count().reset_index()
```

```python
number_of_ratings.head()
#how many review scores does this game name have?
```

```python
number_of_ratings.rename(columns={"Review Score":"number of ratings"}, inplace=True)
```

```python
number_of_ratings.head()
```

```python
finished_set = ratings_games.merge(number_of_ratings, on="name")
```

```python
finished_set.head()
```

```python
finished_set.sample(10)
```

```python
finished_set
```

```python
finished_set.shape
```

```python
games_pivot_table = finished_set.pivot_table(columns='User ID', index='name', values='Review Score')
```

```python
games_pivot_table
```

```python
games_pivot_table.shape
```

```python
#Make Nan 0
games_pivot_table.fillna(0, inplace=True)
```

```python
games_pivot_table.head(23)
#games_pivot_table['name'][0]
```

```python
#0 is ignored in distance calculation
from scipy.sparse import csr_matrix
```

```python
games_sparse_mx = csr_matrix(games_pivot_table)
```

```python
games_sparse_mx
```

```python
#Training Segment: k nearest neighbor
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
#Pass compressed matrix to train
model.fit(games_sparse_mx)
```

```python
#distance, reccomendation = model.kneighbors(games_pivot_table.iloc[52, :].values.reshape(1, -1), n_neighbors=10)
```

```python
def reccomend_games(game_name):
    game_id = np.where(games_pivot_table.index == game_name)[0][0]
    distance, reccomendation = model.kneighbors(games_pivot_table.iloc[game_id, :].values.reshape(1, -1), n_neighbors=10)

    for i in range(len(reccomendation)):
        games = games_pivot_table.index[reccomendation[i]]
        for j in games:
            print(j)
```
